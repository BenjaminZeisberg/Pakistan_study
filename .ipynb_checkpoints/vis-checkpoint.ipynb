```python
# helper functions


# Plots a Polygon to pyplot `ax`
def plot_polygon(ax, poly, **kwargs):
    path = Path.make_compound_path(
        Path(np.asarray(poly.exterior.coords)[:, :2]),
        *[Path(np.asarray(ring.coords)[:, :2]) for ring in poly.interiors])

    patch = PathPatch(path, **kwargs)
    collection = PatchCollection([patch], **kwargs)
    
    ax.add_collection(collection, autolim=True)
    ax.autoscale_view()
    return collection
```

```python
# let's visualize multiple plots per one districts to show spatial variation
columns = df_rain.columns 

# Selecting tehsils that have 5 datapoints in them
six_points = list(columns[:23])
df_rain_subset = df_rain[df_rain[six_points].notnull().all(axis=1)]

tehsils_plot = df_rain_subset.teh.unique()
def return_polygon(teh):
    row = df_adm3.loc[df_adm3['shapeName'] == teh]
    return row['geometry']

tehsils = list(map(return_polygon, tehsils_plot))
tehsils_dic = dict(zip(tehsils_plot, tehsils))
yaz = df_rain_subset.loc[df_rain_subset['teh'] == 'yazman']
```

```python
# Plots a Polygon to pyplot `ax`
def plot_point(ax, lat, lon, **kwargs):
    ax.set_aspect('equal')
    ax.scatter(lat, lon)
```

```python
def get_point_lat_lon(teh, point):
    lat, lon = (df_rain.loc[df_rain['teh'] == teh][f'Point{point}_lat'].iloc[0], df_rain.loc[df_rain['teh'] == teh][f'Point{point}_lon'].iloc[0])
    return lat, lon

def value_point_rain(teh, point):
    date = (df_rain.loc[df_rain['teh'] == teh]['date'].iloc[-12:])
    average = (df_rain.loc[df_rain['teh'] == teh][f'Point{point}_average'].iloc[-12:])
    std = (df_rain.loc[df_rain['teh'] == teh][f'Point{point}_std'].iloc[-12:])
    return date, average, std

def value_point_temp(teh, point):
    date = (df_temp.loc[df_temp['teh'] == teh]['date'].iloc[-12:])
    average = (df_temp.loc[df_temp['teh'] == teh][f'Point{point}_average'].iloc[-12:])
    std = (df_temp.loc[df_temp['teh'] == teh][f'Point{point}_std'].iloc[-12:])
    return date, average, std

date, average, std = value_point_rain('fort abbas', 0)
```

```python
polygon1 = tehsils_dic['fort abbas'].iloc[0]
polygon2 = tehsils_dic['chaubara'].iloc[0]
polygon3 = tehsils_dic['rojhan'].iloc[0]
names = ['fort abbas', 'chaubara', 'rojhan']
colors = ['red', 'green', 'blue', 'grey', 'magenta']
polygon = [polygon1, polygon2, polygon3]

fig, ax = plt.subplots(3, 3, figsize=(20,20))
for i in range(0, 3):
    plot_polygon(ax[0][i], polygon[i], facecolor='lightblue')
    plot_polygon(ax[0][i], polygon[i], facecolor='lightblue')
    plot_polygon(ax[0][i], polygon[i], facecolor='lightblue')
    
    # Plot titles for first row
    ax[0][i].title.set_text(f'{names[i]}')
    
    for j in range(0, 5):
        point = get_point_lat_lon(names[i], j)
        ax[0][i].scatter(point[1], point[0], s=65, color=colors[j])
        
    
    for n in range(0, 5):
        date, average, std = value_point_rain(names[i], n)
        ax[1][i].plot(date, average, label=f'Point {n}',  color=colors[n])
        ax[1][i].tick_params(axis='x', rotation=45)
        ax[1][i].legend()
        ax[1][i].title.set_text(f'{names[i]} spatial variation of rain')
        
    for n in range(0, 5):
        date, average, std = value_point_temp(names[i], n)
        ax[2][i].plot(date, average, label=f'Point {n}', color=colors[n])
        ax[2][i].tick_params(axis='x', rotation=45)
        ax[2][i].legend()
        ax[2][i].title.set_text(f'{names[i]} spatial variation of temperature')
        
plt.legend()
plt.show()
```

```python
points = ['Point0_lat', 'Point0_lon', 'Point1_lat', 'Point1_lon', 'Point2_lat', 'Point2_lon', 'Point3_lat', 'Point3_lon']
# Extract only the lat-lon columns
lat_lon_cols = [col for col in df.columns if ('_lat' in col) or ('_lon' in col)]

def plot_tehsil(points, teh, tehsils_dic, lat_lon_cols):
    if points == 0:
        points = len(lat_lon_cols)
    else: 
        points = 2 * points
    df = df_rain_subset.loc[df_rain_subset['teh'] == teh]
    # Create a base map
    m = folium.Map(location=[df['Point0_lat'].mean(), df['Point0_lon'].mean()], zoom_start=6)
    
    # plot the polygon
    polygon = tehsils_dic[teh].iloc[0]
    folium.GeoJson(polygon).add_to(m)


    # add points
    for i in range(0, len(lat_lon_cols[:points]), 2):  # iterate by 2 because you have pairs of lat and lon
        col_lat = lat_lon_cols[i]
        col_lon = lat_lon_cols[i+1]

        lat = df[col_lat].unique()
        lon = df[col_lon].unique()
        folium.Marker([lat, lon], tooltip=f"{lat_lon_cols[i]}:{lat}, {lat_lon_cols[i+1]}:{lon}").add_to(m)
    
    return m
    
m_test = plot_tehsil(0,'yazman', tehsils_dic, lat_lon_cols)
m_test
```
